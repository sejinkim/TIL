```python
# 방금 만든 것
class Customer:
    def __init__(self, name):
        self.name = name
        self.items = []
        self.prices = []

    def add_item(self, item_name, price):
        self.items.append(item_name)
        self.prices.append(price)

    def total(self):
        return sum(self.prices)

class GSCustomer(Customer):
    def __init__(self, name, membership):
        # 여기서 부모의 __init__도 실행해야 하고
        # membership도 추가로 저장해야 해요
        self.membership = membership
```

```python
gs = GSCustomer("김철수", "골드")
gs.add_item("사과", 1500)
print(gs.name)   # 어떻게 될까요?
```

```python
class GSCustomer(Customer):
    def __init__(self, name, membership):
        super().__init__(name)
        self.membership = membership
    
    def show_receipt(self):
        print(f'=== GS마트 영수증 ({self.membership} 회원) ===')
        for x, y in zip(self.items, self.prices):
            print(f'{x}: {y}원')
        print(f'총액: {self.total()}원')

gs = GSCustomer("김철수", "골드")
gs.add_item("사과", 1500)
gs.add_item("우유", 2000)
gs.show_receipt()
```
