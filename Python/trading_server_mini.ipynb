```python
# Week 1 학습 리뷰를 위한 Mini Project

from abc import ABC, abstractmethod
from dataclasses import dataclass
from enum import Enum
from typing import Optional

# 조건 1 - OrderType Enum 만들기
'''
아래 세 가지 주문 타입을 가진 OrderType을 만들어요:

- LIMIT = "00" (지정가)
- MARKET = "01" (시장가)
- CANCEL = "02" (취소)
'''
class OrderType(str, Enum):
    LIMIT   = "00"      # 지정가
    MARKET  = "01"      # 시장가
    CANCEL  = "02"      # 취소

# 조건 2 - Order dataclass 만들기
'''
아래 속성을 가진 Order를 만들어요:
- symbol: str — 종목코드
- quantity: int — 수량
- price: int — 가격
- order_type: OrderType — 주문 타입 (기본값: OrderType.LIMIT)

그리고 __post_init__에서 아래를 검증해요:
- quantity가 0 이하면 ValueError 발생
- price가 0 미만이면 ValueError 발생
'''

@dataclass  #인자 이름과 속성 이름을 반복하여 쓸 필요가 없게 해 줌.
class Order:
    symbol: str
    quantity: int
    price: Optional[int] = None
    order_type: OrderType = OrderType.LIMIT

    def __post_init__(self):    # __init__ 실행 직후 자동으로 호출됨. 주로 데이터 검증에 씀.
        if self.quantity <= 0:
            raise ValueError("quantity must be > 0")

        if self.price is not None and self.price < 0:
            raise ValueError("price must be >= 0")

# 조건 3 - BaseAccount 추상 클래스 만들기
'''
아래 두 개의 추상 메서드를 가진 BaseAccount를 만들어요:
- submit_order(order: Order) — 주문 제출
- show_summary() — 계좌 요약 출력

그리고 모든 계좌 공통으로 쓰는 아래 기능도 만들어요:
- __init__(self, account_no: str) — 계좌번호, 빈 주문 리스트 초기화
- get_order_count() -> int — 제출된 주문 수 반환
'''

# @dataclass - 제거 : BaseAccount는 데이터를 담는 클래스가 아니라 "공통 규칙을 정의하는 추상 클래스. ABC와 @dataclass를 같이 쓰면 충돌이 생김"
class BaseAccount(ABC):     # @abstractmethod만 있고 ABC를 상속받지 않으면, 자식 클래스가 추상 메서드를 구현하지 않아도 오류가 발생하지 않음. @abstractmethod의 강제력은 ABC에서 나옴.
    def __init__(self, account_no: str):
        self.account_no = account_no
        self.order_list = []
    
    def get_order_count(self) -> int:
        return len(self.order_list)
    
    @abstractmethod
    def submit_order(self, order: Order):
        pass
    
    @abstractmethod
    def show_summary(self):
        pass

# 조건 4 - DemoAccount 클래스 만들기
'''
BaseAccount를 상속받아요. 모의투자 계좌예요.

- __init__에서 super().__init__() 호출 후 balance: int (잔고) 추가
- submit_order(order) 구현:
    - 주문을 리스트에 추가
    - MARKET 주문이면 "[모의] 시장가 주문 제출: {symbol} {quantity}주" 출력
    - LIMIT 주문이면 "[모의] 지정가 주문 제출: {symbol} {quantity}주 @ {price}원" 출력

- show_summary() 구현:
    - 계좌번호, 잔고, 총 주문 수 출력
'''
class DemoAccount(BaseAccount):
    def __init__(self, account_no, balance: int): 
        super().__init__(account_no)
        self.balance = balance

    def submit_order(self, Order):
        self.order_list.append(Order)
        # self.balance -= Order.quantity * Order.price: 시장가를 모르니 여기서는 pass

        if Order.order_type == OrderType.MARKET:
            print(f'[모의] 시장가 주문 제출: {Order.symbol} {Order.quantity}주')

        if Order.order_type == OrderType.LIMIT:
            print(f'[모의] 지정가 주문 제출: {Order.symbol} {Order.quantity}주 @ {Order.price}원' )
    
    def show_summary(self):
        print('=== 계좌 요약===')
        print(f'계좌번호: {self.account_no}')
        print(f'잔고: {self.balance:,}원')
        print(f'총 주문 수: {self.get_order_count()}건')



# 조건 5 - 실행 코드
'''
아래 결과가 나오돌고 실행 코드를 작성해요:

[모의] 지정가 주문 제출: 005930 10주 @ 71000원
[모의] 시장가 주문 제출: 000660 5주
[모의] 지정가 주문 제출: 005380 3주 @ 45000원

=== 계좌 요약 ===
계좌번호: 12345678
잔고: 10,000,000원
총 주문 수: 3건

'''

account = DemoAccount("12345678", 1000000)
```

```python
first_order = Order("005930", 10, 71000, OrderType.LIMIT)
account.submit_order(first_order)
```

```python
second_order = Order("000660", 5, order_type=OrderType.MARKET)
account.submit_order(second_order)
```

```python
third_order = Order("005380", 3, 45000, OrderType.LIMIT)
account.submit_order(third_order)
```

```python
account.show_summary()
```
