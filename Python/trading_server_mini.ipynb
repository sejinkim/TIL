```python
# Week 1 학습 리뷰를 위한 Mini Project

from abc import abstractmethod
from dataclasses import dataclass
from enum import Enum

# 조건 1 - OrderType Enum 만들기
'''
아래 세 가지 주문 타입을 가진 OrderType을 만들어요:

- LIMIT = "00" (지정가)
- MARKET = "01" (시장가)
- CANCEL = "02" (취소)
'''
class OrderType(str, Enum):
    LIMIT   = "00"      # 지정가
    MARKET  = "01"      # 시장가
    CANCEL  = "02"      # 취소

# 조건 2 - Order dataclass 만들기
'''
아래 속성을 가진 Order를 만들어요:
- symbol: str — 종목코드
- quantity: int — 수량
- price: int — 가격
- order_type: OrderType — 주문 타입 (기본값: OrderType.LIMIT)

그리고 __post_init__에서 아래를 검증해요:
- quantity가 0 이하면 ValueError 발생
- price가 0 미만이면 ValueError 발생
'''

@dataclass  #인자 이름과 속성 이름을 반복하여 쓸 필요가 없게 해 줌.
class Order:
    symbol: str
    quantity: int
    price: int
    order_type: OrderType = OrderType.LIMIT

    def __post_init__(self):    # __init__ 실행 직후 자동으로 호출됨. 주로 데이터 검증에 씀.
        if self.quantity <= 0 or self.price < 0:
            raise ValueError

# 조건 3 - BaseAccount 추상 클래스 만들기
'''
아래 두 개의 추상 메서드를 가진 BaseAccount를 만들어요:
- submit_order(order: Order) — 주문 제출
- show_summary() — 계좌 요약 출력

그리고 모든 계좌 공통으로 쓰는 아래 기능도 만들어요:
- __init__(self, account_no: str) — 계좌번호, 빈 주문 리스트 초기화
- get_order_count() -> int — 제출된 주문 수 반환
'''

@dataclass
class BaseAccount:
    account_no: str
    
    def get_order_count() -> int:
        pass
    
    @abstractmethod
    def submit_order(order: Order):
        pass
    
    @abstractmethod
    def show_summary():
        pass

# 조건 4 - DemoAccount 클래스 만들기
'''
BaseAccount를 상속받아요. 모의투자 계좌예요.

- __init__에서 super().__init__() 호출 후 balance: int (잔고) 추가
- submit_order(order) 구현:
    - 주문을 리스트에 추가
    - MARKET 주문이면 "[모의] 시장가 주문 제출: {symbol} {quantity}주" 출력
    - LIMIT 주문이면 "[모의] 지정가 주문 제출: {symbol} {quantity}주 @ {price}원" 출력

- show_summary() 구현:
    - 계좌번호, 잔고, 총 주문 수 출력
'''
class DemoAccount(BaseAccount):
    def __init__(self, account_no, balance: int): 
        super().__init__(account_no)
        self.balalce = balance

    def submit_order(Order):
        order_list = []
        order_list.append(Order)

        if Order.order_type == OrderType.MARKET:
            print(f'[모의] 시장가 주문 제출: {Order.symbol} {Order.quantity}주')

        if Order.order_type == OrderType.LIMIT:
            print(f'[모의] 시장가 주문 제출: {Order.symbol} {Order.quantity}주 @ {Order.price}원' )
    
    def show_summary():
        print('=== 계좌 요약===')
        print(f'잔고: {self.balance}원')
        print(f'총 주문 수: {len(order_list)}건')



# 조건 5 - 실행 코드
'''
아래 결과가 나오돌고 실행 코드를 작성해요:

[모의] 지정가 주문 제출: 005930 10주 @ 71000원
[모의] 시장가 주문 제출: 000660 5주
[모의] 지정가 주문 제출: 005380 3주 @ 45000원

=== 계좌 요약 ===
계좌번호: 12345678
잔고: 10,000,000원
총 주문 수: 3건

'''
```
